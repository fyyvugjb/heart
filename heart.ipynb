```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

# Features of the dataset

**age** (int64): Age of the individual.

**sex** (int64): Sex of the individual (1 = Male, 0 = Female).

**cp** (int64): Chest pain type (categorical).

**trestbps** (int64): Resting blood pressure (mm Hg).

**chol** (int64): Serum cholesterol (mg/dl).

**fbs** (int64): Fasting blood sugar (> 120 mg/dl, 1 = True, 0 = False).

**restecg** (int64): Resting electrocardiographic results (categorical).

**thalach** (int64): Maximum heart rate achieved.

**exang** (int64): Exercise-induced angina (1 = Yes, 0 = No).

**oldpeak** (float64): ST depression induced by exercise relative to rest.

**slope** (int64): The slope of the peak exercise ST segment (categorical).

**ca** (int64): Number of major vessels colored by fluoroscopy (0–3).

**thal** (int64): Thalassemia (categorical).

**target** (int64): Presence of heart disease (1 = Yes, 0 = No).

# Outlier Detection
Outliers were analyzed using the Interquartile Range (IQR) method and were potentially removed:




```python
df.plot(kind='box',subplots=True,layout=(3,5),figsize=(13,15),color='blue',patch_artist=True)
plt.show()
```

```python
Q1=np.quantile(df, 0.25)
Q2=np.quantile(df, 0.50)
Q3=np.quantile(df, 0.75)
Q1, Q2, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
def remove_outliers_iqr(log_data, column):
     # Calculate Q1 (25th percentile) and Q3 (75th percentile) for the column
     Q1 = log_data[column].quantile(0.25)
     Q3 = log_data[column].quantile(0.75)

     # Calculate the Interquartile Range (IQR)
     IQR = Q3 - Q1

     # Define the lower and upper bounds for detecting outliers
     lower_bound = Q1 - 1.5  * IQR
     upper_bound = Q3 + 1.5 * IQR

     # Filter out the outliers from the DataFrame
     log_data_filtered = log_data[(log_data[column] >= lower_bound) & (log_data[column] <= upper_bound)]
     return log_data_filtered
     for i in ["cp","thalach","exang","oldpeak","slope","ca","thal","target"]:
      log_data = remove_outliers_iqr(log_data, i)
```
